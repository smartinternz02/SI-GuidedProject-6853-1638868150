```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sea
```

```python
data=sea.load_dataset("titanic")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop("deck",axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
data[data["embarked"].isnull()]
```

```python
data.drop([61,829],inplace=True)
```

```python
data.isnull().sum()
```

```python
data["age"].fillna(data["age"].median(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sea.boxplot(data["fare"])
```

```python
q1=data["fare"].describe()["25%"]
q3=data["fare"].describe()["75%"]
```

```python
iqr=q3-q1
```

```python
lb=q1-(1.5*iqr)
```

```python
ub=q3+(1.5*iqr)
```

```python
data[data["fare"]<lb].shape[0]
```

```python
data[data["fare"]>ub].shape[0]
```

```python
data["fare"].shape[0]
```

```python
(data[data["fare"]>ub].shape[0]/data["fare"].shape[0])*100
```

```python
sea.boxplot(data["age"])
```

```python
q1=data["age"].describe()["25%"]
q3=data["age"].describe()["75%"]
iqr=q3-q1
lb=q1-(1.5*iqr)
ub=q3+(1.5*iqr)
data[data["age"]<lb].shape[0]
data[data["age"]>ub].shape[0]
data["age"].shape[0]
(data[data["age"]>ub].shape[0]/data["age"].shape[0])*100
```

```python
sea.barplot(data=data,x="embark_town",y="survived")
plot.xlabel("Embark_town")
plot.ylabel("Survived")
plot.title("Survived people based on embark town")
plot.show()
```

```python
plot.scatter(data = data, x= "class",y = "fare")
plot.xlabel("Class")
plot.ylabel("Fare")
plot.title("Ticket Fare on the basis of different classes")
plot.show()
```

```python
who = data["who"].value_counts()
```

```python
plot.pie(who.values,labels=who.index,autopct="%.2f%%")
plot.title("Number of Men, Women and Child")
plot.show()
```

```python
plot.hist(data["age"])
plot.xlabel("Age")
plot.ylabel("Number of People")
plot.title("Histogram on age")
plot.show()
```

```python
sex = data["sex"].value_counts()
```

```python
plot.pie(sex.values,labels=sex.index, autopct = "%.2f%%",colors=["#008E98","#10DFEE"],pctdistance= 0.7)
plot.show()
```

```python
sea.violinplot(x = "sex",y = "age",hue = "survived",data = data,split = True)
plot.show()
```

```python
sea.barplot(data = data,x="pclass", y='survived')
plot.xlabel("Pclass")
plot.ylabel("Survived")
plot.title("Number of survived people based on pclass")
plot.show()
```
